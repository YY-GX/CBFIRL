```python
import numpy as np
import pickle
import matplotlib.pyplot as plt
```

```python
with open('data/debug_data/reward_s_16obs_stop.pkl', 'rb') as f:
    reward_trajs = pickle.load(f)
```

```python
for reward_ls in reward_trajs[:10]:
    plt.plot(range(len(reward_ls)), reward_ls)
    plt.show()
```

```python
ls = [1, 2, 3, 4]
max_lgth = 10
np.concatenate([ls, np.repeat(ls[-1], max_lgth - len(ls))])
```

```python
with open('data/debug_data/reward_s_16obs_stop_400epoch.pkl', 'rb') as f:
    reward_trajs = pickle.load(f)
max_lgth = max([len(reward_ls) for reward_ls in reward_trajs])
ls = np.array([np.concatenate([reward_ls, np.repeat(reward_ls[-1], max_lgth - len(reward_ls))]) for reward_ls in reward_trajs])
plt.plot(range(max_lgth), np.mean(ls, axis=0))
plt.errorbar(range(max_lgth), np.mean(ls, axis=0), np.std(ls, axis=0), linestyle='None', marker='^')
plt.xlabel('timestep')
plt.ylabel('reward')
plt.title('r(T(s, \pi(s))) case')
plt.savefig('figures/r2')
plt.show()
```
